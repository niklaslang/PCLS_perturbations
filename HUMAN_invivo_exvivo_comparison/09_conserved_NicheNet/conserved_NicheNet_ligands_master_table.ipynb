```r
results_dir = '/home/niklas/projects/niche_environments_FIBROSIS/HUMAN_invivo_exvivo_comparison/01_data/ASK_joint/NicheNet/'
```

```r
invivo_dir = '/home/niklas/projects/niche_environments_FIBROSIS/HUMAN_invivo/01_data/NicheNet_outputs/ligand_activities_FC_CC/'
exvivo_dir = '/home/niklas/projects/niche_environments_FIBROSIS/HUMAN_exvivo/01_data/NicheNet_outputs/ligand_activities_FC_CC/'
```

```r
cell_type_labels <- c('Airway_Epithelium','Alveolar_Epithelium',
               'capillary_EC', 'vascular_EC', 'lymphatic_EC',
               'Fibroblasts','SMC_Pericytes',
               'Macrophages','DC','Mast_cells',
               'B_cells','Plasma_cells','T_cells','NK_cells')
```

```r
all_tabs <- data.frame(matrix(0, 0, 4), stringsAsFactors = F, check.names = F)
colnames(all_tabs) <- c('ligand', 'pearson', 'rank_score', 'receiver')
```

```r
pearson_thresh = 0.05
top = 25
```

```r
## loop over EX VIVO receiver cell types
for(ct in cell_type_labels){
    
    tab <- read.csv(paste0(exvivo_dir, 'nichenet_ligand_act_', ct, '.csv'), stringsAsFactors = F)
    tab <- tab[(tab$pearson > pearson_thresh), c("test_ligand", "pearson")]
    tab <- tab[1:min(top, nrow(tab)), ]
    tab$rank_score <- seq(nrow(tab))
    tab$receiver <- ct
    all_tabs <- rbind(all_tabs, tab)
    
}
all_tabs <- reshape(all_tabs, idvar = 'test_ligand', timevar = 'receiver', direction = 'wide')
```

```r
all_tabs <- all_tabs[which(!is.na(all_tabs$test_ligand)),]
col_order <- paste(c(rep('pearson', length(cell_type_labels)), rep('rank_score', length(cell_type_labels))), cell_type_labels, sep = '.')
rownames(all_tabs) <- all_tabs$test_ligand
all_tabs <- all_tabs[, col_order]
```

```r
all_tabs
```

```r
## save results
write.csv(all_tabs, paste0(results_dir, 'exvivo_MASTER_ligands_top', top, '.csv'))
```
