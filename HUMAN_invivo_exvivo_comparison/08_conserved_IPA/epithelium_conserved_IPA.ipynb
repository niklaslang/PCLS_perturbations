```python
pathways = ['Notch Signaling','Acute Phase Response Signaling',
            'PAK Signaling','iNOS Signaling','p53 Signaling','WNT/β-catenin Signaling',
            'Leukocyte Extravasation Signaling','TREM1 Signaling','Interferon Signaling',
            'PI3K/AKT Signaling','Inhibition of Matrix Metalloproteases','Apoptosis Signaling',
            'Superpathway of Cholesterol Biosynthesis', 'PPAR Signaling','Tumor Microenvironment Pathway',
            'JAK/STAT Signaling','Pyroptosis Signaling Pathway',
            'Neuroinflammation Signaling Pathway','Pulmonary Fibrosis Idiopathic Signaling Pathway','IL-13 Signaling Pathway',
            'Cell Cycle: G2/M DNA Damage Checkpoint Regulation','IL-13 Signaling Pathway','Growth Hormone Signaling',
            'ERK/MAPK Signaling','BAG2 Signaling Pathway',
           ]
```

```python
ipa = ipa[ipa['IPA Canonical Pathways'].isin(pathways)]
```

```python
conserved_invivo = ipa[['IPA Canonical Pathways','Invivo']]
conserved_invivo = conserved_invivo.sort_values(by = 'Invivo', ascending = False)
conserved_invivo.rename(columns = {'Invivo':'z-score'}, inplace = True)
conserved_invivo['condition'] = 'invivo'
conserved_exvivo = ipa[['IPA Canonical Pathways','Exvivo']]
conserved_exvivo.rename(columns = {'Exvivo':'z-score'}, inplace = True)
conserved_exvivo['condition'] = 'exvivo'
```

```python
conserved_induced = pd.concat([conserved_invivo, conserved_exvivo], axis=0)
```

```python
conserved_induced
```

```python
sb.barplot(data = conserved_induced,
           x = 'z-score',
           y = 'IPA Canonical Pathways',
           hue = 'condition',
           palette = ['#7209b7','#f72585'],
           ax = plt.figure(figsize=(4, 8), dpi = 100).gca()
          ).set(title='Top conserved IPA pathways',
                xlabel = 'IPA z-score',
                ylabel = 'IPA pathway'
               )
plt.savefig(fig_dir + 'barplot_alveolar_epithelium_IPA_conserved_pathway.pdf')
plt.show()
```
