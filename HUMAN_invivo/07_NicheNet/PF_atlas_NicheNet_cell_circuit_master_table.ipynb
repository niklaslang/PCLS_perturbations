```r
suppressPackageStartupMessages(library(tidyverse))
```

```r
results_dir = '/home/niklas/projects/niche_environments_FIBROSIS/HUMAN_invivo/01_data/NicheNet_outputs/ligand_activities_cell_circuit_pairwise/INVIVO_'
```

```r
cell_type_names <- c('Aberrant Basaloid','Myofibroblasts','ectopic EC', 'SMC/Pericytes_ILD', 'Profibrotic Macrophages')
cell_type_labels <- c('Aberrant_Basaloid','Myofibroblasts','ectopic_EC', 'SMC_Pericytes_ILD', 'Profibrotic_Macrophages')
```

```r
all_tabs <- data.frame(matrix(0, 0, 4), stringsAsFactors = F, check.names = F)
colnames(all_tabs) <- c('ligand', 'pearson', 'rank_score', 'receiver')
```

```r
pearson_thresh = 0.00
top = 50
```

```r
## loop over receiver cell types
for(receiver in cell_type_labels){
    
    ## loop over sender cell type
    for(sender in cell_type_labels){
        tab <- read.csv(paste0(results_dir, 'nichenet_ligand_act_', sender, '_to_', receiver, '.csv'), stringsAsFactors = F)
        tab <- tab[(tab$pearson > pearson_thresh), c("test_ligand", "pearson")]
        tab <- tab[1:min(top, nrow(tab)), ]
        tab$rank_score <- seq(nrow(tab))
        tab$receiver <- paste0(sender, '_to_', receiver)
        all_tabs <- rbind(all_tabs, tab)
    }  
    
}
all_tabs <- reshape(all_tabs, idvar = 'test_ligand', timevar = 'receiver', direction = 'wide')
```

```r
all_tabs <- all_tabs[which(!is.na(all_tabs$test_ligand)),]
col_order <- c(paste0('pearson.', rep(cell_type_labels,5),'_to_', rep(cell_type_labels, each = 5)), paste0('rank_score.', rep(cell_type_labels,5),'_to_', rep(cell_type_labels, each = 5)))
rownames(all_tabs) <- all_tabs$test_ligand
all_tabs <- all_tabs[, col_order]
```
